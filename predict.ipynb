```python
import numpy as np
from PIL import Image
import tensorflow as tf
import matplotlib.pyplot as plt 
```

```python
#loading the model
model = tf.keras.models.load_model('mnist_model.keras')

def predicting_image(img_path):
    img = Image.open(img_path).convert('L')
    img = img.resize((28,28))
    img = np.array(img)

    img_array = img.astype('float32') / 255.0
    img_input = img_array.reshape(1,28,28)

    #predict
    predict = model.predict(img_input)
    predicted_dig = np.argmax(predict)
    confid = np.max(predict) * 100

    #plotting
    plt.imshow(img_array , cmap='grey')
    plt.title(f"Predicted {predicted_dig} , ({confid:.2f}%)")
    plt.show()

    return predicted_dig
```

```python
path = r'test\3.png'
result = predicting_image(path)
```
